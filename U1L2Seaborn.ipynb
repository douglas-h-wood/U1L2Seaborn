```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats, integrate
%matplotlib inline
crashes = sns.load_dataset("car_crashes")
attention = sns.load_dataset("attention")
anscombe = sns.load_dataset("anscombe")
BikesJuly2017 = pd.read_csv('20167-citibike-tripdata.csv')
```

```python
#Histogram with Kernel Density Estimation#
sns.distplot(crashes.speeding)
```

```python
#Just the histogram, number of bins automatically generated, rugplot at bottom#
sns.distplot(crashes.speeding, kde=False, rug=True)
```

```python
#Just the KDE and rugplot#
sns.distplot(crashes.speeding, hist=False, rug=True)
```

```python
#Fitting a parametric distribution to the data#
sns.distplot(crashes.speeding, kde=False, fit=stats.gamma)
```

```python
#Scatterplot for two variables#
sns.jointplot(x="speeding", y="alcohol", data=crashes)
```

```python
#Hexbin plot#
sns.jointplot(x="speeding", y="alcohol", kind="hex", color="k",data=crashes)
```

```python
#Kernel Density Estimation with Two Variables#
sns.jointplot(x="speeding", y="alcohol", data=crashes, kind="kde")
```

```python
#stripplot - a scatterplot where one of the variables is categorical#
sns.stripplot(x="attention", y="score", data=attention)
```

```python
#stripplot with jitter to increase visibility#
sns.stripplot(x="attention", y="score", data=attention,jitter=True)
```

```python
#swarmplot to increase visibility a different way#
sns.swarmplot(x="attention", y="score", data=attention)
```

```python
#Useful to put categorical variable on vertical axis if category name is long or many categories#
sns.swarmplot(y="attention", x="score", data=attention)
```

```python
#swarmplot of the dataset anscombe - categories arranged in order appear in dataframe by default#
sns.swarmplot(x="x", y="y", hue = "dataset",data=anscombe)
```

```python
#boxplot#
sns.boxplot(x="attention", y="score",data=attention)
```

```python
#violionplot combines boxplot with kernel density estimation#
sns.violinplot(x="attention", y="score", data=attention)
```
